**Load libraries**

```python
# data load on google colab
import io
from google.colab import files

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn import metrics
```

**Load data**

```python
uploaded = files.upload()
```

```python
titanic = pd.read_csv(io.BytesIO(uploaded['logistic_data_titanic.csv']))
# titanic['constant'] = 1
titanic = pd.get_dummies(titanic, columns=['Sex'])
titanic.head()
```

**Process data**

```python
# Convert gender column to binary variables
# titanic = pd.get_dummies(titanic, columns=['Sex'])

y = titanic['Survived']
# X = titanic.loc[:, ['constant', 'Age', 'Sex_male', 'Pclass']]
X = titanic.loc[:, ['Age', 'Sex_male', 'Pclass']]
```

**Check classes balance of the dependent variable**


```python
y.value_counts()
```

**Split the data into train and test datasets**


```python
#set seed for reproducibility of results
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=4321
)
```

**Run logistic regression**

```python
# Fit the model
log_reg = sm.Logit(y_train, X_train).fit()

# Print the summary table
print(log_reg.summary())
```

**Confusion matrix**

```python
# Predictions on the test dataset
y_predicted = log_reg.predict(X_test)
binary_prediction = np.where(y_predicted >= 0.5, 1, 0)

# Confusion matrix
print("Confusion Matrix : \n", metrics.confusion_matrix(y_test, binary_prediction))

# Performance metrics
print('Test accuracy = ', metrics.accuracy_score(y_test, binary_prediction))
print('Test precision = ', metrics.precision_score(y_test, binary_prediction))
```
